# Componentes principales

```python
data_selected = df_wine.iloc[:, 3:5].values
pd.DataFrame(data_selected)
```

```python
pd.DataFrame(data_selected).describe()
```

# Clustering Jerarquico

```python
import scipy.cluster.hierarchy as shc

plt.figure(figsize=(10, 7))
plt.title("Dendogramas de vinos")
dend = shc.dendrogram(shc.linkage(data_selected, method='ward'))
```

```python
from sklearn.cluster import AgglomerativeClustering

cluster = AgglomerativeClustering(n_clusters=3, affinity='euclidean', linkage='ward')
cluster.fit_predict(data_selected)
```

```python
plt.figure(figsize=(10, 7))
plt.scatter(data_selected[:,0], data_selected[:,1], c=cluster.labels_,cmap='rainbow')
plt.xlabel("Ash_Alcanity")
plt.ylabel("Magnesium")
plt.show()
```

La extraccion de los componentes principales arroja que el magnesio y las cenizas alcalinas son los 2 principales componentes para agruparlos, el corte en el dendograma que puede ser entre 100 a 125 arroja 3 clusters.
