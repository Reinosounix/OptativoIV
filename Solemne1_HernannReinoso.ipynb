```python
#se considera no esencial para la prediccion del salario el número de serie, la junta de educación secundaria, 
#la junta de educación superior y el estado de colocacion.

df = df.drop(['sl_no','ssc_b','hsc_b','status'], axis='columns')
df
```

```python
for feature in df.columns:     
    print('Total de valores nulos de', feature, '=', df[feature].isna().sum())
```
